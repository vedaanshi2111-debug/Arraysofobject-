```python
#step 1;  import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# step 2; load iris dataset
iris = load_iris()
x = iris.data
y = iris.target
```

```python
#step 3 ; load Iris Dataset
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Step 4: Create KNN Model
k = 3   # You can change K value
knn = KNeighborsClassifier(n_neighbors=k)
```

```python

# Step 5: Train the Model
knn.fit(X_train, y_train)
```

```python

# Step 6: Make Predictions
y_pred = knn.predict(X_test)
```

```python
# Step 7: Evaluate Model
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#implementation of LogisticRegression
from sklearn.linear_model import LogisticRegression
# Step 1: Create Logistic Regression Model
model = LogisticRegression(max_iter=200)

# Step 2: Train the Model
model.fit(X_train, y_train)

# Step 3: Make Predictions
y_pred = model.predict(X_test)

# Step 4: Evaluate Model
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#implementation of Stochastic Gradient Descent (SGD) Classifier
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
# Step 1: Feature Scaling (IMPORTANT for SGD)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 2: Create SGD Classifier
model = SGDClassifier(loss='log_loss', max_iter=1000, random_state=42)

# Step 3: Train the Model
model.fit(X_train, y_train)

# Step 4: Make Predictions
y_pred = model.predict(X_test)

# Step 5: Evaluate Model
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#implementation of Kernel Support Vector Machine (SVM)
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
# Step 1: Feature Scaling (IMPORTANT for SVM)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 2: Create Kernel SVM Model (RBF Kernel)
model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)

# Step 3: Train the Model
model.fit(X_train, y_train)

# Step 4: Make Predictions
y_pred = model.predict(X_test)

# Step 5: Evaluate Model
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#implementation of Naive Bayes
from sklearn.naive_bayes import GaussianNB

# Step 1: Create Naive Bayes Model
model = GaussianNB()

# Step 2: Train the Model
model.fit(X_train, y_train)

# Step 3: Make Predictions
y_pred = model.predict(X_test)

# Step 4: Evaluate Model
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#implementation of Decision Tree Classification
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Step 1: Create Decision Tree Model
model = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)

# Step 2: Train the Model
model.fit(X_train, y_train)

# Step 3: Make Predictions
y_pred = model.predict(X_test)

# Step 4: Evaluate Model
print("Accuracy:", accuracy_score(y_test, y_pred))

# Step 5: Visualize the Decision Tree
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=iris.feature_names,
          class_names=iris.target_names, filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```
